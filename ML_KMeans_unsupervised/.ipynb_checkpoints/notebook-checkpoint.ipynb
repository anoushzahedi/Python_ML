You have been asked to support a team of researchers who have been collecting data about penguins in Antartica! The data is available in csv-Format as `penguins.csv`

**Origin of this data** : Data were collected and made available by Dr. Kristen Gorman and the Palmer Station, Antarctica LTER, a member of the Long Term Ecological Research Network.

**The dataset consists of 5 columns.**

Column | Description
--- | ---
culmen_length_mm | culmen length (mm)
culmen_depth_mm | culmen depth (mm)
flipper_length_mm | flipper length (mm)
body_mass_g | body mass (g)
sex | penguin sex

Unfortunately, they have not been able to record the species of penguin, but they know that there are **at least three** species that are native to the region: **Adelie**, **Chinstrap**, and **Gentoo**.  Your task is to apply your data science skills to help them identify groups in the dataset!

```python
import os
os.environ['OMP_NUM_THREADS'] = '2'
```

```python
# Libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
penguins_df = pd.read_csv("penguins.csv")
print(penguins_df.isnull().sum())
print('*'*30)
print(penguins_df.dtypes)
print('*'*30)
# changing sex to numerical: male=0, female=1
penguins_df['sex'] = penguins_df['sex'].map({'MALE':0, 'FEMALE':1})
print(penguins_df.head())

# plotting variables
fig, ax = plt.subplots(3,2)
ax = ax.ravel()
for i, col in enumerate(penguins_df.columns):
    sns.histplot(penguins_df[col], ax=ax[i], kde=True)
    ax[i].set_title(f"Distribution of {col}")
fig.delaxes(ax[5])
plt.tight_layout()
plt.show()
```

```python
# Libs
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline

# searching for the best n_clusters using silhouette scores
score = []
ks = np.arange(2,10)
for k in ks:
    pipeline = make_pipeline(StandardScaler(), 
                            KMeans(n_clusters=k, random_state = 123, n_init=10))
    labels = pipeline.fit_predict(penguins_df)
    X_scaled = pipeline[0].transform(penguins_df)
    score_tmp = silhouette_score(X_scaled, labels)
    score.append(score_tmp)


# final results
score_df =pd.DataFrame(list(zip(ks, score)), columns = ['k', 'silhouette_score'])
best_k, best_score = score_df.loc[score_df['silhouette_score'].idxmax(), ['k', 'silhouette_score']]
print('*'*30)
print(f"the best k:{best_k}, with the score of {best_score:.3f}")
print('*'*30)

# plotting for understanding the results 
plt.plot(score_df['k'], score_df['silhouette_score'])
plt.annotate(f'Best: k={best_k}, score={best_score:3f}',
             xy=(best_k, best_score),
             xytext=(best_k - 2, best_score - 0.15),
             arrowprops=dict(arrowstyle='->', color='red'))
sns
plt.show()
```

```python
# re-running the best estimator
pipeline = make_pipeline(StandardScaler(),
                        KMeans(n_clusters=int(best_k), n_init=10))
penguins_df['labels'] = pipeline.fit_predict(penguins_df)

# checking the relationship between sex and lables
print(pd.crosstab(penguins_df['labels'],penguins_df['sex']))

# producing stat_penguins
stat_penguins =  penguins_df.drop('sex', axis=1).groupby('labels').mean()
print(stat_penguins.head(6))
```
